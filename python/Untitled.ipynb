```python
import pandas as pd
```

```python
lei_df = pd.read_csv("../output/fuzzy_first_run.txt", dtype="object", sep="|")
len(lei_df)
```

```python
lei_df.columns
```

```python
lei_df = lei_df[['inst_id',
                'nic_name',
                'panel_city',
                'panel_street',
                'nic_lei',
                'gleif_lei',
                'gleif_name',
                'gleif_city',
                'gleif_street',
                'city_score',
                'name_score',
                'street_score']].copy()
```

```python
lei_df["total_score"] = lei_df.apply(lambda x: int(x.city_score) + int(x.name_score) + int(x.street_score), axis=1)
lei_df.sort_values(by="total_score", ascending=False, inplace=True)
lei_df
```
